**Perbedaan dari list, tuple, set dan dictionary dalam python :**<p>
-**List** adalah tipe data yang paling serbaguna dalam bahasa pemrograman Python. List ditulis sebagai daftar nilai yang dipisahkan koma antara tanda kurung siku. <p>
-**Tuple** adalah tipe data pada Python yang tidak dapat diubah. Artinya, sekali Tuple dibuat, maka isinya tidak dapat diubah lagi. <p>
-**Set** adalah tipe data dalam Python yang digunakan untuk menyimpan beberapa item dalam satu variabel.<p>
-**Dictionary** adalah tipe data yang memungkinkan untuk menyimpan pasangan kunci-nilai. Dictionary dapat didefinisikan dengan menggunakan tanda kurung kurawal dan pasangan kunci-nilai yang dipisahkan oleh koma.

```python
dictionary_ku= {
    "Nama"          : "Gollum",
    "Kelas"         : "17",
    "Tanggal lahir" : "15 Oktober 2003",
    "No. Hp"      : "085723456789",
    "Alamat"      : "Magelang",
    "Hobi"          : ['Mabar','Main','Tidur']
}

for key, value in dictionary_ku.items():
   if key == "Hobi" :
    print(key,":", ",".join(value))
   else :
    print(key,":",value)
```
