```python
phi = 3.14

def luas_lingkaran():
  r=float(input("Masukkan nilai jari-jari lingkaran: "))
  return phi * r ** 2

def luas_persegi():
  sisi=float(input("Masukkan nilai sisi persegi: "))
  return sisi ** 2

def luas_persegipanjang():
  panjang=float(input("Masukkan nilai panjang persegi panjang: "))
  lebar =float(input("Masukkan nilai lebar persegi panjang: "))
  return panjang*lebar
```

```python
import mtk

print("Luas lingkaran",mtk.luas_lingkaran())
print("Luas persegi",mtk.luas_persegi())
print("Luas persegi panjang",mtk.luas_persegipanjang())
```

```python
import math

number = 5
factorial = math.factorial(number)
print(f"The factorial of {number} is {factorial}")

number = 16
sqrt = math.sqrt(number)
print(f"The square root of {number} is {sqrt}")

pi = math.pi
print(f"The value of pi is {pi}")

angle_rad = math.pi / 2
sin_val = math.sin(angle_rad)
print(f"The sine of {angle_rad} radians is {sin_val}")

num1 = 48
num2 = 18
gcd = math.gcd(num1, num2)
print(f"The greatest common divisor of {num1} and {num2} is {gcd}")
```
