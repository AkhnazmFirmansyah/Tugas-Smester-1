```python
class atk:
  def __init__(self, nama ,stok, hargasatuan,):
    self.nama = nama
    self.stok = stok
    self.hargasatuan = hargasatuan
    self.hargatotal = hargasatuan*stok

  def set_nama(self, nama):
    self.nama = nama

  def set_stok(self, stok):
    self.stok = stok

  def set_hargasatuan(self, hargasatuan):
    self.hargasatuan = hargasatuan

  def set_hargatotal(self, hargatotal):
    self.hargatotal = self.hargasatuan * self.stok

  def get_nama(self):
    return self.nama

  def get_stok(self):
    return self.stok

  def get_hargasatuan(self):
    return self.hargasatuan

  def get_hargatotal(self):
    return self.hargatotal

Bolpoint1 = atk("Bolpoint", 10, 2000)
Pensil1 = atk("Pensil", 10, 1000)
Penghapus = atk("Penghapus", 10 , 500)

print("Nama", Bolpoint1.get_nama(), "Stok", Bolpoint1.get_stok(), "Harga Satuan", Bolpoint1.get_hargasatuan(), "Harga Total", Bolpoint1.get_hargatotal())
print("Nama", Pensil1.get_nama(), "Stok", Pensil1.get_stok(), "Harga Satuan", Pensil1.get_hargasatuan(), "Harga Total", Pensil1.get_hargatotal())
print("Nama", Penghapus.get_nama(), "Stok", Penghapus.get_stok(), "Harga Satuan", Penghapus.get_hargasatuan(), "Harga Total", Penghapus.get_hargatotal())

hargatotal = Bolpoint1.get_hargatotal() + Pensil1.get_hargatotal() + Penghapus.get_hargatotal()
print("Harga Total", hargatotal)
```
