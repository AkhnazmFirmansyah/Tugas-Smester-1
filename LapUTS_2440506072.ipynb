**Weekly Exercise 1**

```python
npm = input("Masukkan NPM Anda : ")
stringnpm = str(npm)

genap = 0
ganjil = 0

for i in range(len(stringnpm)):
    if int(stringnpm[i]) % 2 == 0:
        genap += 1
    else:
        ganjil += 1

print("Jumlah Angka Genap adalah : ", genap)
print("Jumlah Angka Ganjil adalah :", ganjil)
```

**Weekly Exercise 2**

```python
baris = int(input("Masukkan Jumlah Baris : "))
for i in range(1, baris + 1):
    for j in range(1, i+1):
        print(i*j, end=" ")
    print()
```
