**Weekly Exercise 1**

```python
def lingkaran():
  try:
    phi = 3.14
    r = int(input("Jari-jari lingkaran: "))
    luas = phi*r*r
    print("Luas lingkaran adalah: ", luas)
  except ValueError:
    print("Masukkan angka!")
def persegi():
  try:
    sisi = int(input("Sisi persegi: "))
    luas = sisi*sisi
    print("Luas persegi adalah: ", luas)
  except ValueError:
    print("Masukkan angka!")
def segitiga():
  try:
    alas = int(input("Alas segitiga: "))
    tinggi = int(input("Tinggi segitiga: "))
    luas = alas*tinggi/2
    print("Luas segitiga adalah: ", luas)
  except ValueError:
    print("Masukkan angka!")

def function():
  while True:
    print("Pilih perhitungan: ")
    print("1. Lingkaran")
    print("2. Persegi")
    print("3. Segitiga")

    choice = input("Masukkan pilihan anda (1/2/3): ")
    if choice in ('1', '2', '3'):
        if choice == '1':
            lingkaran()
        elif choice == '2':
            persegi()
        elif choice == '3':
            segitiga()
        break
    else:
        print("Input salah")
function()
```

```python
try:
  npm = input("Masukkan NPM Anda : ")
  stringnpm = str(npm)

  genap = 0
  ganjil = 0

  for i in range(len(stringnpm)):
     if int(stringnpm[i]) % 2 == 0:
        genap += 1
     else:
        ganjil += 1

  print("Jumlah Angka Genap adalah : ", genap)
  print("Jumlah Angka Ganjil adalah :", ganjil)
except ValueError:
  print("Masukkan NPM dengan benar!")
```
