**Weekly Exercise 1**

```python
def faktorial(x):
     if x == 0 :
      return 1
     else :
      return x * faktorial(x-1)

x = int(input("Masukkan angka yang akan difaktorialkan: "))
faktorial = faktorial(x)
print(x, "!", "=", faktorial)
```

**Weekkly Exercise 2**

```python
def lingkaran():
    phi = 3.14
    r = int(input("Jari-jari lingkaran: "))
    luas = phi*r*r
    print("Luas lingkaran adalah: ", luas)
def persegi():
    sisi = int(input("Sisi persegi: "))
    luas = sisi*sisi
    print("Luas persegi adalah: ", luas)
def segitiga():
    alas = int(input("Alas segitiga: "))
    tinggi = int(input("Tinggi segitiga: "))
    luas = alas*tinggi/2
    print("Luas segitiga adalah: ", luas)

def function():
  while True:
    print("Pilih perhitungan: ")
    print("1. Lingkaran")
    print("2. Persegi")
    print("3. Segitiga")

    choice = input("Masukkan pilihan anda (1/2/3): ")
    if choice in ('1', '2', '3'):
        if choice == '1':
            lingkaran()
        elif choice == '2':
            persegi()
        elif choice == '3':
            segitiga()
        break
    else:
        print("Input salah")
function()
```
